```python
def preprocess(data):
    """This method preprocesses sensor's data by minifying the data"""
    interval_window=20
    initial=0
    data2=[]
    count=0
    for i in range(10, data.acc_X.count()+1, interval_window):
        
        record={
                #mean
                'acc_X': data[initial:i].acc_X.mean(), 'acc_Y': data[initial:i].acc_Y.mean(),
                #'acc_Z': data[initial:i].acc_Z.mean(),
                #std
                'acc_X_std': data[initial:i].acc_X.std(), 'acc_Y_std': data[initial:i].acc_Y.std(),
                #'acc_Z_std': data[initial:i].acc_Z.std(),
                #max
                'acc_X_max': data[initial:i].acc_X.max(), 'acc_Y_max': data[initial:i].acc_Y.max(),
                #'acc_Z_max': data[initial:i].acc_Z.max(),
                #min
                'acc_X_min': data[initial:i].acc_X.min(), 'acc_Y_min': data[initial:i].acc_Y.min(),
                #'acc_Z_min': data[initial:i].acc_Z.min(),
                #mad(median average deviation)
                'acc_X_mad': data[initial:i].acc_X.mad(), 'acc_Y_mad': data[initial:i].acc_Y.mad(),
                #'acc_Z_mad': data[initial:i].acc_Z.mad(),
                #iqr-inter quantile range
                'acc_X_iqr': (data[initial:i].acc_X.quantile(0.75)-data[initial:i].acc_X.quantile(0.25)), 
                'acc_Y_iqr': (data[initial:i].acc_Y.quantile(0.75)-data[initial:i].acc_Y.quantile(0.25)),
                #'acc_Z_iqr': (data[initial:i].acc_Z.quantile(0.75)-data[initial:i].acc_Z.quantile(0.25)),
                #quantile 1
                'acc_X_q1': data[initial:i].acc_X.quantile(0.25), 'acc_Y_q1': data[initial:i].acc_Y.quantile(0.25),
                #'acc_Z_q1': data[initial:i].acc_Z.quantile(0.25),
                #quantile 2
                'acc_X_q2': data[initial:i].acc_X.quantile(0.5), 'acc_Y_q2': data[initial:i].acc_Y.quantile(0.5),
                #'acc_Z_q2': data[initial:i].acc_Z.quantile(0.5),
                #quantile 3
                'acc_X_q3': data[initial:i].acc_X.quantile(0.75), 'acc_Y_q3': data[initial:i].acc_Y.quantile(0.75),
                #'acc_Z_q3': data[initial:i].acc_Z.quantile(0.75),
                #skewness
                'acc_X_skew': data[initial:i].acc_X.skew(), 'acc_Y_skew': data[initial:i].acc_Y.skew(),
                #'acc_Z_skew': data[initial:i].acc_Z.skew(),
                #kurtosis
                'acc_X_kurt': data[initial:i].acc_X.kurt(), 'acc_Y_kurt': data[initial:i].acc_Y.kurt(),
                #'acc_Z_kurt': data[initial:i].acc_Z.kurt(),
                #Energy
                'acc_X_energy': np.divide(np.sum(np.square(data[initial:i].acc_X)), interval_window), 
                'acc_Y_energy': np.divide(np.sum(np.square(data[initial:i].acc_Y)), interval_window),
                #'acc_Z_energy': np.divide(np.sum(np.square(data[initial:i].acc_Z)), interval_window),
                }
        data2.append(record)
        initial=initial+interval_window
        count=count+1;
        
    return data2
        
    
```

```python
%time data2=preprocess(data)
```

```python
d=pd.DataFrame(data=data2)
d.describe()
```
