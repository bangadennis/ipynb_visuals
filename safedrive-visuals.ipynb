## Safedrive collected data Visualizations

## Vehicle-Test003

```python
#import dependencies
import matplotlib.pyplot as plt
import numpy as np
from firebase import firebase
import pandas as pd

%matplotlib inline
```

```python
#Reading data in json format
test003loc=pd.read_json("safedrive-app-test003-export-loc.json")
test003loc=test003loc.T
test003loc["limitspeed"]=80
```

```python
##setting up the plot
fig, axes=plt.subplots(figsize=(20, 8))

axes.set_title("Speed against Time -Vehicle Test003")
axes.set_xlabel("Time")
axes.set_ylabel("Speed(Km/hrs)")
#line speedlimit
axes.plot(test003loc.timestamp, test003loc.limitspeed, "r--")
#plot
axes.plot(test003loc.timestamp, test003loc.speed)
```

### Accelerations vs time

```python
#Reading data in json format
test003acc=pd.read_json("safedrive-app-test003-export-acc.json")
test003acc=test003acc.T
```

```python
#plot
fig, axes=plt.subplots(figsize=(20, 10))

axes.set_title("Accelerations against Time -Vehicle Test003")
axes.set_xlabel("Time")
axes.set_ylabel("Acceleration")
#line speedlimit
axes.plot(test003acc.timestamp, test003acc.acc_X, "b-")
axes.plot(test003acc.timestamp, test003acc.acc_Y, "r-")
#axes.plot(test003acc.timestamp, test003acc.acc_Z, "g-")

axes.legend(["X", "Y", "Z"])
```
