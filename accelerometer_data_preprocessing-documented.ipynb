## Data Preprocessing & Feature Creation

```python
#Preprocess Data
def preprocess(data, interval_window=10):
    """This method preprocesses sensor's data by minifying the data"""
    initial=0
    data_processed=[]
    count=0
    
    if data.empty:
        return data_processed

    for i in range(interval_window, data.acc_X.count()+1, interval_window):
        
        record={
                #mean
                #'acc_X': data[initial:i].acc_X.mean(), 
                #'acc_Y': data[initial:i].acc_Y.mean(),
                #'acc_Z': data[initial:i].acc_Z.mean(),
                #std
                #'acc_X_std': data[initial:i].acc_X.std(), 
                #'acc_Y_std': data[initial:i].acc_Y.std(),
                #'acc_Z_std': data[initial:i].acc_Z.std(),
                #max
                #'acc_X_max': data[initial:i].acc_X.max(), 
                #'acc_Y_max': data[initial:i].acc_Y.max(),
                #'acc_Z_max': data[initial:i].acc_Z.max(),
                #min
                #'acc_X_min': data[initial:i].acc_X.min(), 
                #'acc_Y_min': data[initial:i].acc_Y.min(),
                #'acc_Z_min': data[initial:i].acc_Z.min(),
                #mad(median average deviation)
                #'acc_X_mad': data[initial:i].acc_X.mad(), 
                #'acc_Y_mad': data[initial:i].acc_Y.mad(),
                #'acc_Z_mad': data[initial:i].acc_Z.mad(),
                #iqr-inter quantile range
                #'acc_X_iqr': (data[initial:i].acc_X.quantile(0.75)-data[initial:i].acc_X.quantile(0.25)), 
                #'acc_Y_iqr': (data[initial:i].acc_Y.quantile(0.75)-data[initial:i].acc_Y.quantile(0.25)),
                #'acc_Z_iqr': (data[initial:i].acc_Z.quantile(0.75)-data[initial:i].acc_Z.quantile(0.25)),
                #quantile 1
                #'acc_X_q1': data[initial:i].acc_X.quantile(0.25), 'acc_Y_q1': data[initial:i].acc_Y.quantile(0.25),
                #'acc_Z_q1': data[initial:i].acc_Z.quantile(0.25),
                #quantile 2
                #'acc_X_q2': data[initial:i].acc_X.quantile(0.5), 'acc_Y_q2': data[initial:i].acc_Y.quantile(0.5),
                #'acc_Z_q2': data[initial:i].acc_Z.quantile(0.5),
                #quantile 3
                #'acc_X_q3': data[initial:i].acc_X.quantile(0.75), 'acc_Y_q3': data[initial:i].acc_Y.quantile(0.75),
                #'acc_Z_q3': data[initial:i].acc_Z.quantile(0.75),
                #skewness
                #'acc_X_skew': data[initial:i].acc_X.skew(), 
                'acc_Y_skew': data[initial:i].acc_Y.skew(),
                #'acc_Z_skew': data[initial:i].acc_Z.skew(),
                #kurtosis
                #'acc_X_kurt': data[initial:i].acc_X.kurt(), 
                #'acc_Y_kurt': data[initial:i].acc_Y.kurt(),
                #'acc_Z_kurt': data[initial:i].acc_Z.kurt(),
                #Energy
                'acc_X_energy': np.divide(np.sum(np.square(data[initial:i].acc_X)), interval_window), 
                'acc_Y_energy': np.divide(np.sum(np.square(data[initial:i].acc_Y)), interval_window),
                'acc_Z_energy': np.divide(np.sum(np.square(data[initial:i].acc_Z)), interval_window),
                #correlation XY,XZ, YZ
                #'acc_corr_XY': data.acc_X[initial:i].corr(data.acc_Y[initial:i]), 
                #'acc_corr_XZ': data.acc_X[initial:i].corr(data.acc_Z[initial:i]),
                #'acc_corr_YZ': data.acc_Y[initial:i].corr(data.acc_Z[initial:i]),
                }
        data_processed.append(record)
        initial=initial+interval_window
        count=count+1;
        
    return data_processed
        
    
```

## Data for Training the model

```python
datamodel2=pd.read_csv("Desktop/safedrive/model-safedrive-attributes.csv")
```

```python
y_values=pd.read_csv("Desktop/safedrive/model-safedrive-class.csv")
```

## Splitting data into training and test sets

```python
 X_train, X_test, y_train, y_test = train_test_split(datamodel2.values, y_values.target, random_state=0)
```

## KNN

```python
model2=neighbors.KNeighborsClassifier(n_neighbors=5)
model2.fit(X_train, y_train)
```

### KNN-Confusion Matrix

```python
y_predict=model2.predict(X_test)
metrics.confusion_matrix(y_test, y_predict)
```

```python
m=metrics.classification_report(y_test, y_predict)
print(m)
```

## SVM

```python
#SVC- Support Vector Machine 
model2=svm.SVC()
model2.fit(X_train, y_train)
```

### SVM-Confusion Matrix and Classification Report

```python
y_predict=model2.predict(X_test)
metrics.confusion_matrix(y_test, y_predict)
```

```python
m=metrics.classification_report(y_test, y_predict)
print(m)
```
